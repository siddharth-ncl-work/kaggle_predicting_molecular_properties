## Train

```
norm_cols=['x_0','y_0','z_0','x_1','y_1','z_1','mulliken_charge_0','mulliken_charge_1','distance',
           'rc_A', 'rc_B', 'rc_C', 'mu', 'alpha', 'homo', 'lumo',
           'gap', 'r2', 'zpve', 'U0', 'U', 'H', 'G', 'Cv', 'freqs_min',
           'freqs_max', 'freqs_mean', 'linear', 'mulliken_min', 'mulliken_max',
           'mulliken_mean', 'mulliken_atom_0', 'mulliken_atom_1']
```

```
from sklearn.preprocessing import StandardScaler
```

```
std_scaler=StandardScaler()
train_norm[norm_cols]=std_scaler.fit_transform(train_norm[norm_cols])
```

```
train_norm.describe(include='all').T
```

## Test

```
test_norm=pd.read_csv('test_feat_eng_v2_orig_my_ext_not_std.csv')
test_norm.head()
```

```
test_norm.shape
```

```
test_norm[norm_cols]=std_scaler.transform(test_norm[norm_cols])
```
